```python
import pandas as pd

df = pd.DataFrame(pd.read_csv("/kaggle/input/the-global-hunger-index/global-hunger-index.csv"))

df
```

```python
keep_col=['Entity','Year','Global Hunger Index (2021)']
df=df[keep_col]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.pairplot(df,hue='Entity')
plt.show()
```
