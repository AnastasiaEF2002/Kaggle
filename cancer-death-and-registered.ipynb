```python
lesion_type_dict = {
    'regs': 1,
    'deaths': 2,
}
df['type'] = df.type.map(lesion_type_dict)
df
```

```python
lesion_type_dict = {
    'AllSex': 0,
    'Female': 1,
    'Male': 2,
}
df['sex'] = df.sex.map(lesion_type_dict)
df
```

```python
df.corr()
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.datasets import make_regression
X = df.drop(['numbers'],axis= 1)
y=df['numbers']
regr = RandomForestRegressor(max_depth=10, random_state=0)
regr.fit(X, y)
print(regr.predict([[3000, 2, 0]]))
```
