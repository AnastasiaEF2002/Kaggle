```python
import pandas as pd

df = pd.DataFrame(pd.read_csv("/kaggle/input/metallica-studio-tracks-with-spotify-data/MetallicaAllTracks.csv"))

df
```

```python
keep_col = ['Year','Time_Signature','Danceability','Energy','Key','Loudness','Mode','Speechiness','Acousticness','Instrumentalness','Liveness','Valence','Tempo','Popularity']
f = df[keep_col]
f.corr()
```

```python
f=f.dropna()
f.corr()
```

```python
f
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn import preprocessing
X = f.drop(['Popularity'],axis= 1)
y = f['Popularity']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.9, random_state=42)
model = LinearRegression()
model.fit(X_train,y_train)
predictions = model.predict(X_test)
print(
  'mean_squared_error : ', mean_squared_error(y_test, predictions))
print(
  'mean_absolute_error : ', mean_absolute_error(y_test, predictions))
```

```python
model.intercept_
```

```python
import numpy as np

coefficients = model.coef_

rounded_coefficients = [round(coef, 3) for coef in coefficients]

print(rounded_coefficients)
```
