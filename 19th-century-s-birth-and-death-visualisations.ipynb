```python
import pandas as pd

df = pd.DataFrame(pd.read_csv("/kaggle/input/survey-data/yearly_deaths_by_clinic.csv"))

df
```

```python
sns.lineplot(x="year", y="deaths", data=df)
 
# setting the title using Matplotlib
plt.title('Title using Matplotlib Function')
 
plt.show()
```
