```python
keep_col=['Year','Any opioid death rates (CDC WONDER)','Cocaine overdose death rates (CDC WONDER)','Heroin overdose death rates (CDC WONDER)','Synthetic opioids death rates (CDC WONDER)','Prescription Opioids death rates (US CDC WONDER)']
df=df[keep_col]
df.corr()
```

```python
from sklearn.ensemble import RandomForestRegressor

X = df.drop(['Any opioid death rates (CDC WONDER)'],axis= 1)
y=df['Any opioid death rates (CDC WONDER)']
regr = RandomForestRegressor(max_depth=4, random_state=0)
regr.fit(X, y)
print(regr.predict([[2030,6,4,18,5]]))
```
