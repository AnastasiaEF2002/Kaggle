```python
df = open("/kaggle/input/mortality-rate-in-the-usa-by-gender-area-cause/USRegionalMortality.csv", "r") 
  
df = ''.join([i for i in df])  
  
# search and replace the contents 
df = df.replace("Urban", "0")  
df = df.replace("Rural", "1")  
df = df.replace("Female", "0")  
df = df.replace("Male", "1")
df=df.replace("Heart disease","1")
df=df.replace("Cancer","2")
df=df.replace("Lower respiratory","3")
df=df.replace("Unintentional injuries","4")
df=df.replace("Cerebrovascular diseases","5")
df=df.replace("Alzheimers","6")
df=df.replace("Diabetes","7")
df=df.replace("Flu and pneumonia","8")
df=df.replace("Suicide","9")
df=df.replace("Nephritis","10")
df=df.replace("HHS Region 01","1")
df=df.replace("HHS Region 02","2")
df=df.replace("HHS Region 03","3")
df=df.replace("HHS Region 04","4")
df=df.replace("HHS Region 05","5")
df=df.replace("HHS Region 06","6")
df=df.replace("HHS Region 07","7")
df=df.replace("HHS Region 08","8")
df=df.replace("HHS Region 09","9")
df=df.replace("HHS Region 10","10")

# output.csv is the output file opened in write mode 
x = open("output.csv","w") 
  
# all the replaced text is written in the output.csv file 
x.writelines(df) 
x.close()
```
